Mengimport library pandas untuk membaca dataset

```python
import pandas as pd

atribute = ['bahan','cara_masak','rasa','nama_makan']
data = pd.read_csv("/content/DatasetsAI.csv")
```

Merubah data yang tadinya berupa tipe data string dan menjadi biner

```python
makanan = pd.get_dummies(data[['bahan','cara_masak','rasa']])
```

Mengimport modul sklearn decision tree

```python
from sklearn import tree
```

Kita akan menggunakan algoritma decision tree clasifier untuk memprediksi nama makanan berdasarkan bahan cara masak dan rasa makanan, Kita juga akan menentukan target sebagai hasil untuk prediksi

```python
hasil = tree.DecisionTreeClassifier(criterion='entropy')
data_hasil = hasil.fit(makanan,data['nama_makan'])
```

Membuat data latih dan data uji dengan menggunakan kode berikut

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(makanan, data['nama_makan'], test_size=0.25)
y_pred = hasil.predict(X_test)
```

Menetukan akurasi dari model yang telah kita buat

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)

print("Akurasi:", accuracy)
```
